```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Drop the specified columns from the DataFrame
columns_to_drop = ['Unnamed: 0', 'Date', 'CustomerID', 'Transaction_ID', 'Product_SKU', 'Product_Description']
online_shopping = online_shopping.drop(columns=columns_to_drop)

# Display the first few rows of the cleaned DataFrame
print(online_shopping.head())
```

```python
# Create an additional column (Total_price)

online_shopping['Total_price'] = online_shopping['Avg_Price'] * online_shopping['Quantity'] + online_shopping['Delivery_Charges']
```

```python
# Create an additional column (Total_Spend)

online_shopping['Total_Spend'] = online_shopping['Offline_Spend']+online_shopping['Online_Spend']
```

```python
# Drop duplicates from the entire DataFrame
online_shopping = online_shopping.drop_duplicates()
online_shopping.duplicated().sum()
```

# DATA VISUALIZATION

```python
# Calculate the distribution of male and female customers
gender_distribution = online_shopping['Gender'].value_counts()

# Create a pie chart for the gender distribution
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', explode=[0.08, 0])

# Ensure the pie chart is a perfect circle
plt.axis('equal')

# Set the plot title
plt.title('Distribution of Male and Female Customers')

# Set Seaborn style
sns.set(style='white')

# Display the plot
plt.show()
```

```python
# Calculate the total amount spent by male and female customers
gender_totals = online_shopping.groupby('Gender')['Total_price'].sum()

# Set seaborn style
sns.set(style='white')

# Plot the pie chart
plt.pie(gender_totals, labels=gender_totals.index, autopct="%1.1f%%", shadow=True, explode=(0.1, 0))

# Make sure the pie chart is a circle
plt.axis('equal')

# Add a title
plt.title('Total Price Spent by Male / Female Customer in (%)')

# Display the pie chart
plt.show()
```

```python
gender_total_spent = online_shopping.groupby('Gender')['Total_price'].sum()

# Convert the group totals to a list (male, female) and ensure labels match
# the values order
values = [gender_total_spent['F'], gender_total_spent['M']]

# Set seaborn style to white
sns.set(style='white')

# Create a pie chart with the values and labels
plt.pie(values, labels=['Female', 'Male'], autopct="%1.1f%%", shadow=True, explode=(0.1, 0))

# Ensure the pie chart is a perfect circle
plt.axis('equal')

# Set the title of the plot
plt.title('Price Spent by Male / Female Customer')

# Display the pie chart
plt.show()
```

```python
tenure=online_shopping.Tenure_Months
sns.histplot(tenure,kde=True)
plt.xlabel('Months')
plt.ylabel('Frequency')
plt.title('Tenure Months Frequency of Customers')
```

```python
# Determine the top 20 most frequent customer tenure months
top_tenure_months = online_shopping["Tenure_Months"].value_counts().head(20).index

# Create a count plot of customer tenure in months
plt.figure(figsize=(10, 4))
sns.countplot(x="Tenure_Months", data=online_shopping, order=top_tenure_months, palette="hls")

# Set the title for the plot
plt.title("Customer Tenure in Months")

# Display the plot
plt.show()
```

```python
# Calculate the frequency of each product category
freq_product = online_shopping['Product_Category'].value_counts()

# Calculate the total number of products
total_products = freq_product.sum()

# Calculate the percentage for each category
percentages = (freq_product / total_products) * 100

# Create a swarm plot using seaborn
sns.scatterplot(x=freq_product.index, y=percentages)

# Rotate x-axis tick labels for better readability
plt.xticks(rotation=85)

# Set plot title and labels
plt.title('Distribution of Product Categories')
plt.ylabel('Percentages')

# Display the plot
plt.show()
```

```python
# Plot monthly offline sales
plt.figure(figsize=(10, 4))
online_shopping.groupby("Month")["Offline_Spend"].sum().plot()
plt.grid()
plt.title("Monthly Offline Sales")
plt.xlabel("Month")
plt.ylabel("Sales in Naira")
plt.show()

# Plot monthly online sales
plt.figure(figsize=(10, 4))
online_shopping.groupby("Month")["Online_Spend"].sum().plot(color="red")
plt.title("Monthly Online Sales")
plt.xlabel("Month")
plt.ylabel("Sales in Naira")
plt.show()
```

```python
# Calculate the top 5 product categories by online spend
top_online_categories = online_shopping.groupby("Product_Category")["Online_Spend"].sum().nlargest(5).reset_index()

# Plot the top 5 product categories by online spend
plt.figure(figsize=(10, 4))
sns.lineplot(x="Product_Category", y="Online_Spend", data=top_online_categories, marker='o')
plt.title("Top 5 Categories by Online Spend")
plt.show()

# Calculate the top 5 product categories by offline spend
top_offline_categories = online_shopping.groupby("Product_Category")["Offline_Spend"].sum().nlargest(5).reset_index()

# Plot the top 5 product categories by offline spend
plt.figure(figsize=(10, 4))
sns.lineplot(x="Product_Category", y="Offline_Spend", data=top_offline_categories, marker='o', color="green")
plt.title("Top 5 Categories by Offline Spend")
plt.show()
```
